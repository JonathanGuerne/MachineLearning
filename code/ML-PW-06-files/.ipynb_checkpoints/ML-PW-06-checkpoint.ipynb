### Exercice 1 Classification to predict student admission

##### a. Logistic regression classifier with linear decision boundary

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True)

df = pd.read_csv('./student-dataset-train.csv',names=['x1','x2','y'])
#print(df.head())

x = np.array([np.ones(len(df['x1'])),df['x1'].values,df['x2'].values]).T
y = df['y'].values

#print(x[:10,:])
#print(y[:10])

print(x.shape)
print(y.shape)
```

```python
x1_mean = np.mean(x[:,1])
x2_mean = np.mean(x[:,2])
x1_var = np.std(x[:,1])
x2_var = np.std(x[:,2])

x[:,1] = (x[:,1] - x1_mean) / x1_var
x[:,2] = (x[:,2] - x2_mean) / x2_var
```

```python
def g(z):
    return 1/(1+np.exp(-z))
```

```python
z_value = np.arange(-10,10,0.01)
sigmoid = g(z_value)

plt.title("Sigmoid")
plt.plot(z_value,sigmoid)
plt.show()
```

```python
def hypothesis(X,theta):
    return g(np.dot(X,theta))
```

```python
def objective(X,y,theta):
    epsilon = 1e-6
    h = hypothesis(X,theta)
    N = X.shape[0]
    tmp = y * np.log(h + epsilon) + (1-y) * np.log(1 - h + epsilon)
    return np.sum(tmp)/N
```

```python
def gradientAscent(X,y,alpha,nEpoch):
    N = X.shape[0]
    D = X.shape[1]
    
    J = []
    
    theta = np.zeros(D)

    for i in range(nEpoch):
        h = hypothesis(X,theta)
        
        for d in range(D):
            theta[d] = theta[d] + alpha * (1/N) * ((y - h)*X.T[d]).sum()     
        
        J.append(objective(x,y,theta))

    return theta,J
```

```python
nEpoch = 
alpha = 0.001

theta,J = gradientAscent(x,y,alpha,nEpoch)

print(theta)

Xepoch = np.arange(0,nEpoch)

plt.plot(Xepoch,J)
plt.xlabel("Epoch")
plt.ylabel("Performance value (J)")
plt.show()
```

```python
def plotData_Line(x,y,title):
    x1_0 = []
    x1_1 = []
    x2_0 = []
    x2_1 = []

    for i in np.arange(x.shape[0]):
        if y[i] == 0:
            x1_0.append(x[i,1])
            x2_0.append(x[i,2])
        else:
            x1_1.append(x[i,1])
            x2_1.append(x[i,2])

    plt.scatter(x1_0,x2_0,marker='o',color='green',label='not admitted')
    plt.scatter(x1_1,x2_1,marker='x',color='red',label='admitted')
    plt.xlabel('Exam 1 score $x_1$')
    plt.ylabel('Exam 2 score $x_2$')
    plt.legend(bbox_to_anchor=(1.1, 1))
    axes = plt.gca()
    #axes.set_xlim([20,110])
    #axes.set_ylim([20,110])

    #t0 + t1*x1 + t2*x2 = 0
    #t0 + t1*x1 = -t2*x2
    #-(t0 + t1*x1)/t2 = x2

    x2_borderLine = lambda x1 : -(theta[0] + theta[1]*x1)/theta[2]
    x1_value = np.arange(np.min(x[:,1]),np.max(x[:,1]),0.01)

    plt.plot(x1_value,x2_borderLine(x1_value))

    plt.title(title)

    plt.show()
    
plotData_Line(x,y,"Train set with training result")
```

```python
df = pd.read_csv('./student-dataset-train.csv',names=['x1','x2','y'])

x_train = np.array([np.ones(len(df['x1'])),df['x1'].values,df['x2'].values]).T
y_train = df['y'].values

df = pd.read_csv('./student-dataset-test.csv',names=['x1','x2','y'])

x_test = np.array([np.ones(len(df['x1'])),df['x1'].values,df['x2'].values]).T
y_test = df['y'].values
```

```python
x1_mean = np.mean(x[:,1])
x2_mean = np.mean(x[:,2])
x1_var = np.std(x[:,1])
x2_var = np.std(x[:,2])

x_train[:,1] = (x_train[:,1] - x1_mean) / x1_var
x_train[:,2] = (x_train[:,2] - x2_mean) / x2_var
```

```python
x_test[:,1] = (x_test[:,1] - x1_mean) / x1_var
x_test[:,2] = (x_test[:,2] - x2_mean) / x2_var
```

```python
from sklearn import linear_model

clf = linear_model.SGDClassifier(loss='log',max_iter = 500000, learning_rate='invscaling',eta0 = 0.01)
clf.fit(x_train,y_train)

t = clf.coef_[0]

preds = clf.predict(x_test)
print("Resultat de la classification = {}/{}".format(np.sum(y_test==preds),len(y_test)))
```
