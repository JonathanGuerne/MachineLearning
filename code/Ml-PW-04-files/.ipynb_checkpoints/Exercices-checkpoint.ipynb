## Exercice 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train = None
test = None

np.set_printoptions(suppress=True)

dataset = pd.read_excel("lausanne-appart.xlsx")
dataset = np.array(dataset)
dataset = dataset[:,:3]
    
X_train = dataset[:,0]
Y_train = dataset[:,2]

plt.hist(X_train[:])
plt.show()

plt.hist(Y_train[:])
plt.show()
```

## Exercice2

```python
from numpy.linalg import inv

N = len(X_train)

def theta(_X, _y):
    _Xt = _X.T
    return np.dot(inv(_Xt.dot(_X)),_Xt.dot(_y))

def h(theta0, theta1, x):
    return theta0 + theta1 * x

def J(theta0, theta1, opt_terme=1):
    return 1 / (2 * N) * np.sum(np.power(h(theta0, theta1, X_train) - Y_train, 2)*opt_terme)

X_set = np.ones((N, 2))
X_set[:, 1] = X_train

thetas = theta(X_set, Y_train)

yHat = h(thetas[0], thetas[1], X_train)

plt.scatter(X_train, Y_train)
plt.plot(X_train, yHat, color='red')
plt.show()

print(J(thetas[0], thetas[1]))
```

## Exercice 3

```python
alpha = 0.00001
cost_value = []
numEpoch = 100
thetas = [0, 0]

def J_b(theta0, theta1, opt_terme=1):
    a = h(theta0, theta1, X_train)
    print(a-Y_train)
    return (1 / N) * np.sum(( a - Y_train)
                            * opt_terme)


for i in range(numEpoch):
    print(J_b(thetas[0], thetas[1]))
    
    thetas[0] = thetas[0] - alpha * J_b(thetas[0], thetas[1])
    break
    thetas[1] = thetas[1] - alpha * J_b(thetas[0], thetas[1], X_train)
    cost_value.append(J(thetas[0], thetas[1]))
    
print(thetas)
    
plt.plot(np.arange(numEpoch), cost_value)
plt.show()

yHat = h(thetas[0], thetas[1], X_train)

plt.scatter(X_train, Y_train)
plt.plot(X_train, yHat, color='red')
plt.show()
```

## Exercice 3

```python
from random import randint

alpha = 0.0001
cost_value = []
numEpoch = 1000
thetas = [0, 0]

def J_s(theta0, theta1, X=1):
    rnd_idx = randint(0, N-1)
    
    if not isinstance(X, int):
        X = X[rnd_idx]
        
    return (1 / N) * ((h(theta0, theta1, X_train[rnd_idx]) - Y_train[rnd_idx]) * X)

for i in range(numEpoch):
    thetas[0] = thetas[0] - alpha * J_s(thetas[0], thetas[1])
    thetas[1] = thetas[1] - alpha * J_s(thetas[0], thetas[1], X_train)
    cost_value.append(J(thetas[0], thetas[1]))
    
print(cost_value[-1])
    
plt.plot(np.arange(numEpoch), cost_value)
plt.show()

yHat = h(thetas[0], thetas[1], X_train)

plt.scatter(X_train, Y_train)
plt.plot(X_train, yHat, color='red')
plt.show()
```
